```python
# 실행마다 동일한 결과를 얻기 위해 케라스에 랜덤 시드를 사용하고 텐서플로 연산을 결정적으로 만듭니다. 
import tensorflow as tf

tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()
```

## 2개의 층

```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)) # 은닉층 100개의 뉴런, 입력의크기 지정
dense2 = keras.layers.Dense(10, activation='softmax') # 출력층 10개 뉴런
```

은닉층의 뉴런 개수를 지정하는 특별한 기준은 없지만 적어도 출력층의 뉴련보다는 많이 만들어야 함

## 심층 신경망 만들기

```python
model = keras.Sequential([dense1, dense2]) # 밀집층을가진 모델 생성
```

```python
model.summary() # 층에 대한 유용한 정보 출력
```

## 층을 추가하는 다른 방법

```python
model = keras.Sequential([
    keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'),
    keras.layers.Dense(10, activation='softmax', name='output')
], name='패션 MNIST 모델') # 한 셀로 밀집층,모델 생성
```

```python
model.summary()
```

```python
model = keras.Sequential() # add로 모델 생성
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy') # 컴파일

model.fit(train_scaled, train_target, epochs=5) # 훈련
# 추가된 층이 훈련세트 정확도 향상시킴
```

## 렐루 활성화 함수

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28))) # 직접 1차원으로 펼치지 않고 Flatten 층 사용
model.add(keras.layers.Dense(100, activation='relu')) # 은닉층에 렐루 함수 사용
model.add(keras.layers.Dense(10, activation='softmax')) 
```

```python
model.summary()
```

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0 # 전처리

train_scaled, val_scaled, train_target, val_target = train_test_split( # 검증세트
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')

model.fit(train_scaled, train_target, epochs=5) # 훈련세트 훈련
```

```python
model.evaluate(val_scaled, val_target) # 검증결과>> 은늑층에 렐루함수를 넣었더니 성능 향상(이미지에서 좋은 성능을 냄)
```

## 옵티마이저

```python
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics='accuracy') # 자동으로 SGD객체 생성
```

```python
sgd = keras.optimizers.SGD() # 기본 경사 하강법 옵티마이저 클래스
model.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
sgd = keras.optimizers.SGD(learning_rate=0.1) # 학습률 지정
```

```python
sgd = keras.optimizers.SGD(momentum=0.9, nesterov=True) # 0이상의 값을 지정하여 모멘템 최적화 수행, True로 네스테로프 모멘텀 최적화 수행
```

```python
adagrad = keras.optimizers.Adagrad() # Adagrad 옵티마이저 클래스
model.compile(optimizer=adagrad, loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
rmsprop = keras.optimizers.RMSprop() # RMSprop 옵티마이저 클래스
model.compile(optimizer=rmsprop, loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
model = keras.Sequential() 
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy') # Adan 옵티마이저 클래스

model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```

* 심층 신경망: 2개 이상의 층을 포함한 신경망

* 렐루함수: 이미지 분류 모델의 은닉층에 많이 사용하는 활성화 함수, 시그모이드 함수의 단점 보완

* 옵티마이저: 케라스는 기본적으로 미니배치 경사하강법 사용, 다양한 경사 하강법 알고리즘
  - SGD, 네스테로프 모멘텀, RMSprop, Adam 등

* 활성화함수: 시그모이드, 소프트맥스, 렐루

* TensorFlow
  - add()는 케라스 모델에 층을 추가하는 메서드
  
  - summary()는 케라스 모델의 정보를 출력하는 메서드
    * 모델에 추가된 층의 종류와 순서 등, name으로 이름 지정

  - SGD: 기본 경사 하강법 클래스
  - Adagrad: 옵티마이저 클래스
  - RMSprop: 옵티마이저 클래스
  - Adam: 옵티마이저 클래스
